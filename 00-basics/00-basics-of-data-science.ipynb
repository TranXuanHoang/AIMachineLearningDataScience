```python
# Import Pandas and Matplotlib's Pyplot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
# Read CSV data and populate it using Pandas
df = pd.read_csv('heart-disease.csv')
df
```

```python
# Extract only the first 10 rows (incl. the header row) of the CSV file
df.head(10)
```

```python
# Look into column 'target', count frequency of different values appearing in this column and plot out a bar chart
# Note that, we call matplotlib.clf() to clear the entire current figure (chart) with all its axes
# that might be generated by other cells, but leave the window opened, such that it might be reused for other plots
plt.clf()
targetChart = df.target.value_counts().plot(kind="pie")
targetChart.figure
```

```python
# Look into column 'age' and count frequency of records for each 'age' value, then plot a bar chart
plt.clf()
age = df.age.value_counts()
ageBarChart = age.plot(kind="bar")
ageBarChart.figure
```

```python
# Plot age in a horizontal bar chart.
# The figsize(width, height) is to specify the size of the chart printed out.
plt.clf()
age.plot(kind="barh", figsize=(10,6)).figure
```

```python
# Plot 'age' and 'chol' in a scatter chart. Scatter chart only accept Data Frames to be called with '.plot'.
plt.clf()
df.plot(kind="scatter", x='age', y='chol', figsize=(10,5), color='Red').figure
```

```python
plt.clf()
df.plot(kind="area", figsize=(20,6)).figure
```
